```python
import pandas as pd
import numpy as np

def load_and_clean_data(filepath):
   
    df = pd.read_csv(filepath)
    df_cleaned = df.drop(columns=[col for col in df.columns if 'Unnamed' in col])
    return df_cleaned

def filter_data(df, target_column):
   
    df_filtered = df.dropna(subset=[target_column])
    return df_filtered

def extract_matrices(df, feature_columns, target_column):
    
    matrix_A = df[feature_columns]
    matrix_C = df[[target_column]]
    return matrix_A, matrix_C

def compute_costs(matrix_A, matrix_C):
   
  
    A = matrix_A.values
    C = matrix_C.values
    
    
    A_pseudo_inv = np.linalg.pinv(A)
    
   
    costs = A_pseudo_inv @ C
    return costs

def main():
   
    filepath = "C:/Users/year3/Documents/Purchase_Data.csv"
    
    
    cleaned_data = load_and_clean_data(filepath)
    
    
    filtered_data = filter_data(cleaned_data, 'Payment (Rs)')
    

    feature_columns = ['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']
    target_column = 'Payment (Rs)'
    

    matrix_A, matrix_C = extract_matrices(filtered_data, feature_columns, target_column)
    
    
    product_costs = compute_costs(matrix_A, matrix_C)
    
   
    print("Computed costs of each product:")
    for idx, cost in enumerate(product_costs):
        print(f"Product {feature_columns[idx]}: {cost[0]:.2f}")

if __name__ == "__main__":
    main()
```
